## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_____________________________________________________________________________________________________________________
LEFT JOIN- returns all records from table A (left table) and the matching record from table B (right table). If there is no match in the right table the result is NULL.
RIGHT JOIN- returns all records from table B (right table) and the matching record from table A (left table). if there is no match in the left table the result is NULL.
INNER JOIN- default. It is the intersection between two tables.
OUTER JOIN- It is not suported by sqlite.
